```python
import selenium
from selenium import webdriver
from modules.driver_loader import DriverLoader
import time
import pandas as pd
from selenium.webdriver.common.by import By
from modules import wait_util
import requests
import pandas as pd
from bs4 import BeautifulSoup
from collections import Counter
import numpy as np
```

```python
driver = DriverLoader().get_chrome_driver()
driver.get('https://www.techrankup.com/zh-Hant/smartphone-processors-ranking/')
data = driver.find_elements_by_class_name('xbar')
```

```python
data
```

```python
cpu_rank = []
for i in data:
    if len(i.text) < 2:
        continue
    else:
        cpu_rank.append(i.text)
```

```python
for i in data:
    print(i.text)
```

```python
cpu_rank
```

```python
for i in cpu_rank:
    if len(i) < 4:
        cpu_rank.remove(i)
```

```python
for i in range(len(cpu_rank)):
    cpu_rank[i] = cpu_rank[i].replace(':',' : ')
```

```python
cpu_rank
```

```python
cpu = []
for i in cpu_rank:
    a = i.index(':')
    cpu.append([i[:a-1],i[a+2:]])
```

```python
import csv
with open('cpu.csv','w', newline='',encoding = 'utf-8') as csvfile:
  writer = csv.writer(csvfile)
  writer.writerow(['cpu','rank'])
  for i in cpu:
      writer.writerow(i)
```
